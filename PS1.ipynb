````python
# Your code here. Add cells as needed.
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler


X = housing_df.drop(columns=['price', 'date', 'zipcode'])


feature_names = X.columns
````

````python
X = X.astype(float)
y = housing_df['price'].astype(float)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
````

````python
from sklearn.metrics import r2_score


alpha_value = 5000  
ridge_model = Ridge(alpha=alpha_value)

ridge_model.fit(X_train, y_train)

y_pred = ridge_model.predict(X_test)


r2 = r2_score(y_test, y_pred)
print(f'R^2 Score: {r2:.4f}')
````

````python
import numpy as np

coefficients = ridge_model.coef_

top_features_indices = np.argsort(np.abs(coefficients))[-3:] 

top_features = feature_names[top_features_indices]
top_values = coefficients[top_features_indices]

print('\nTop 3 Most Important Features:')
for feature, value in zip(top_features, top_values):
    print(f'{feature}: {value:.4f}')




import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5, label='Predicted vs Actual')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='dashed', label='Perfect Prediction')

plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Predicted vs Actual Home Prices (Ridge Regression)')
plt.legend()
plt.show()
````
